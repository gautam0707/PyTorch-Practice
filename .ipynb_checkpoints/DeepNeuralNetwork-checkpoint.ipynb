```python
import torch
import numpy as np
```

```python
data = np.loadtxt('./diabetes.csv.gz', delimiter=',', dtype=np.float32)
```

```python
x_data = torch.from_numpy(data[:,:-1])
y_data = torch.from_numpy(data[:,-1].reshape((len(data),1)))
x_data.requires_grad = True
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = torch.nn.Linear(8, 6)
        self.l2 = torch.nn.Linear(6, 4)
        self.l3 = torch.nn.Linear(4, 1)
    
    def forward(self, x):
        out1 = torch.relu(self.l1(x))
        out2 = torch.relu(self.l2(out1))
        y_pred = torch.sigmoid(self.l3(out2))
        return y_pred
```

```python
model = Model()
criterion = torch.nn.BCELoss(reduction='elementwise_mean')
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)
```

```python
for epoch in range(2000):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(epoch, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
